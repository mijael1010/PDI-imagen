```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image = cv2.imread('F:/tiger.jpg')

# Aumentar el brillo
brightness_decrease = 50
darkened_image = cv2.subtract(image, np.ones(image.shape, dtype=np.uint8) * brightness_decrease)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(darkened_image, cv2.COLOR_BGR2RGB))
plt.title('Imagen con Brillo Disminuido')
plt.axis('off')

plt.show()
```
