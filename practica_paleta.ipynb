```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


paleta_todo = [cv2.COLORMAP_AUTUMN, cv2.COLORMAP_BONE, cv2.COLORMAP_COOL, 
               cv2.COLORMAP_HOT, cv2.COLORMAP_HSV, cv2.COLORMAP_JET, 
               cv2.COLORMAP_OCEAN, cv2.COLORMAP_PARULA, cv2.COLORMAP_PINK, 
               cv2.COLORMAP_RAINBOW, cv2.COLORMAP_SPRING, cv2.COLORMAP_SUMMER, 
               cv2.COLORMAP_WINTER]


imagen = cv2.imread('E:/tiger.jpg', cv2.IMREAD_GRAYSCALE)


plt.subplot(4, 4, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')


for i,k in enumerate(paleta_todo):
    color = cv2.applyColorMap(imagen, k)
    plt.subplot(4, 4, i+2)
    plt.imshow(cv2.cvtColor(color, cv2.COLOR_BGR2RGB))
    plt.title(i)
  
plt.show()
```
