```python
import cv2
import matplotlib.pyplot as plt

def histograma(imagen):
    histo = cv2.calcHist([imagen], [0], None, [256], [0, 256])  
    plt.figure(figsize=(10, 5))
    plt.plot(histo, color="red")
    plt.title('Histograma')
    plt.xlabel('Intensidad del pxel')
    plt.ylabel('Frecuencia')
    plt.show()

def umbralizacion_y_histograma(imagen):
   
    _, umbralbi = cv2.threshold(imagen, 100, 255, cv2.THRESH_BINARY)
    _, umbralbi_inv = cv2.threshold(imagen, 100, 255, cv2.THRESH_BINARY_INV)

    # Umbralizacin adaptativa
    umb_adap = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
    umb_adap_gauss = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    # Mostrar las imgenes
    plt.figure(figsize=(15, 10))

    plt.subplot(2, 3, 1)
    plt.imshow(imagen, cmap='gray')
    plt.title('Imagen inicial')

    plt.subplot(2, 3, 2)
    plt.imshow(umbralbi, cmap='gray')
    plt.title('Umbral binario')

    plt.subplot(2, 3, 3)
    plt.imshow(umbralbi_inv, cmap='gray')
    plt.title('Umbral binario inverso')

   

    plt.subplot(2, 3, 4)
    plt.imshow(umb_adap, cmap='gray')
    plt.title('Umbral adaptativo (media)')

    plt.subplot(2, 3, 5)
    plt.imshow(umb_adap_gauss, cmap='gray')
    plt.title('Umbral adaptativo (gaussiano)')

    plt.show()

   
    histograma(imagen)


imagen = cv2.imread('E:/arthas.jpg', 0)


umbralizacion_y_histograma(imagen)

```
