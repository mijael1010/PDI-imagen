```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,5]

img_cancerbgr = cv2.imread(r"F:/imagen1.jpg") #Imagen Trabajadora


imagen = cv2.cvtColor(img_cancerbgr,cv2.COLOR_BGR2RGB)


cortado = imagen[600:800, 650:850, :]


plt.subplot(1, 2, 1)
plt.imshow(imagen)
plt.title('Imagen Original')
plt.axis('on')

plt.subplot(1, 2, 2)
plt.imshow(cortado)
plt.title('imagen 2')
plt.axis('on')

plt.show()
cv2.imwrite('Imagen_cortado.jpg', cortado)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('F:/imagen1.jpg')


brightness_decrease = 50
g_image = cv2.subtract(image, np.ones(image.shape, dtype=np.uint8) * brightness_decrease)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cmap='gray')  
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(g_image, cv2.COLOR_BGR2GRAY), cmap='gray')  
plt.title('Imagen 2')
plt.axis('off')

plt.show()

cv2.imwrite('Imagen_degra.jpg', g_image)
```
